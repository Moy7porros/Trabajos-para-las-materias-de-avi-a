Ejericico de regresion lineal

```python
#separar las caracteristicas de la etiqueta
x = datos_na.drop(["median_house_value"], axis=1)
y = datos_na["median_house_value"]
```

## Separación de variables independientes y dependiente

En este bloque se separan las variables del conjunto de datos en:

- **X (variables independientes)**: contienen las características que se utilizarán para hacer la predicción.
- **y (variable dependiente)**: corresponde al valor objetivo, en este caso `median_house_value`.

Se elimina la columna `median_house_value` de X porque es el valor que el modelo debe predecir.

Este paso es fundamental en Machine Learning, ya que define claramente qué información se usará como entrada y cuál será la salida esperada del modelo.

```python
#separar los datos en dos partes, un conjunto de entrenamiento y otro de pruebas
from sklearn.model_selection import train_test_split
x_ent, x_pru, y_ent, y_pru = train_test_split(x,y, test_size=.2)
```

## División del conjunto de datos

En este bloque se dividen los datos en dos subconjuntos:

- **Datos de entrenamiento (`x_ent`, `y_ent`)**: utilizados para entrenar el modelo.
- **Datos de prueba (`x_pru`, `y_pru`)**: utilizados para evaluar el rendimiento del modelo.

Se utiliza un 80% de los datos para entrenamiento y un 20% para prueba.

Esta técnica permite verificar qué tan bien el modelo puede generalizar a datos nuevos, evitando problemas como el sobreajuste (overfitting).

```python

x_pru.shape
```

x_pru.shape

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

## Creación del modelo de regresión lineal

Aquí se importa y se crea una instancia del modelo de regresión lineal utilizando la librería `sklearn`.

La regresión lineal es un algoritmo de aprendizaje supervisado que busca modelar la relación entre variables mediante una ecuación lineal.

El objetivo del modelo es encontrar la mejor línea que se ajuste a los datos para poder realizar predicciones de valores continuos.

```python
modelo.fit(x_ent, y_ent)
```

## Entrenamiento del modelo

En este bloque, el modelo aprende a partir de los datos de entrenamiento.

El método `.fit()` ajusta el modelo encontrando:
- Los coeficientes (pendientes)
- El intercepto

Estos parámetros permiten construir la ecuación de la recta que mejor representa la relación entre las variables.

Este es el paso donde el modelo "aprende" patrones a partir de los datos.

```python
predicciones = modelo.predict(x_pru)
```

## Generación de predicciones

Aquí se utilizan los datos de prueba (`x_pru`) para generar predicciones con el modelo entrenado.

El método `.predict()` aplica el modelo aprendido para estimar valores de la variable dependiente.

Estas predicciones serán comparadas con los valores reales para evaluar la precisión del modelo.

```python
predicciones
```

## Visualización de predicciones

Se muestran los valores predichos por el modelo.

Estos valores representan las estimaciones del modelo para el precio medio de las casas (`median_house_value`) basadas en las características de entrada.

Este paso permite observar directamente los resultados generados por el modelo.

```python
comparativa = {"Predicion": predicciones, "Valor Real": y_pru}
pd.DataFrame(comparativa)
```

## Comparación entre valores reales y predichos

Se construye un DataFrame que permite comparar:

- **Predicción**: valores estimados por el modelo
- **Valor real**: valores originales del dataset

Esto facilita analizar qué tan cercanas son las predicciones respecto a la realidad, lo cual es clave para evaluar el desempeño del modelo.

```python
#Overfitting sobreajuste
print(modelo.score(x_ent, y_ent))
print(modelo.score(x_pru, y_pru))
```

## Evaluación del modelo y detección de overfitting

Se calcula el coeficiente de determinación (R²) tanto en entrenamiento como en prueba.

- Si el valor en entrenamiento es mucho mayor que en prueba → puede existir **sobreajuste (overfitting)**.
- Si ambos valores son similares → el modelo generaliza correctamente.

El R² indica qué tan bien el modelo explica la variabilidad de los datos (valores cercanos a 1 son mejores).

```python
#Error
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_pru, predicciones)
```

## Cálculo del error cuadrático medio (MSE)

Se calcula el **Mean Squared Error (MSE)**, que mide el promedio de los errores al cuadrado entre valores reales y predichos.

Un valor bajo de MSE indica que las predicciones están cerca de los valores reales.

Este indicador es fundamental para evaluar la precisión del modelo.

```python
mse
```

## Resultado del MSE

Se muestra el valor del error cuadrático medio.

Este valor permite cuantificar el nivel de error del modelo, aunque no está en la misma escala que la variable original.

```python
rmse = np.sqrt(mse)
```

## Cálculo de la raíz del error cuadrático medio (RMSE)

Se calcula el RMSE aplicando la raíz cuadrada al MSE.

El RMSE tiene la ventaja de estar en la misma escala que la variable original, lo que facilita su interpretación.

Un valor menor indica mejor desempeño del modelo.

```python
rmse
```

## Interpretación del RMSE

Se muestra el valor final del RMSE.

Este valor representa el error promedio de las predicciones del modelo en unidades reales (precio de vivienda).

Permite entender qué tan lejos están, en promedio, las predicciones respecto a los valores reales.
