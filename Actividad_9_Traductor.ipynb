## Introducción

En esta actividad se desarrolla un programa en Python que permite traducir palabras del español al inglés utilizando un diccionario.

Los diccionarios en Python son estructuras de datos que almacenan información en pares clave-valor, lo cual los hace ideales para representar traducciones entre idiomas.

Este proyecto tiene como finalidad aplicar conceptos básicos de programación, como funciones, estructuras de datos y control de flujo, además de documentar adecuadamente el proceso en un cuaderno de Jupyter Notebook.

## Objetivo

Desarrollar un traductor simple que permita convertir palabras del español al inglés utilizando un diccionario en Python, mostrando al menos 10 traducciones y aplicando buenas prácticas de programación.

```python
# Diccionario de traducción (mínimo 10 palabras)
diccionario = {
    "hola": "hello",
    "mundo": "world",
    "perro": "dog",
    "gato": "cat",
    "casa": "house",
    "escuela": "school",
    "comida": "food",
    "agua": "water",
    "libro": "book",
    "mesa": "table"
}
```

El diccionario es una estructura clave-valor donde cada palabra en español (clave) tiene su equivalente en inglés (valor).

Esto permite realizar búsquedas rápidas y eficientes para traducir palabras.

### Función de traducción

Se define una función que recibe una palabra en español y devuelve su traducción en inglés.

```python
def traducir(palabra):
    # Convertir a minúsculas para evitar errores
    palabra = palabra.lower()

    # Buscar en el diccionario
    return diccionario.get(palabra, "Palabra no encontrada")
```

La función `traducir()` realiza lo siguiente:

- Convierte la palabra a minúsculas para evitar errores de coincidencia.
- Utiliza el método `.get()` del diccionario para buscar la palabra.
- Si la palabra no existe, devuelve "Palabra no encontrada".

El método `.get()` es recomendable porque evita errores en caso de que la clave no exista, devolviendo un valor por defecto.

### Pruebas del traductor

A continuación se muestran ejemplos de traducción de palabras.

```python
# Lista de palabras a traducir
palabras = ["hola", "mundo", "perro", "gato", "casa", "escuela", "comida", "agua", "libro", "mesa"]

for palabra in palabras:
    print(f"{palabra} -> {traducir(palabra)}")
```

### Traducción interactiva

Se permite al usuario ingresar palabras para traducirlas dinámicamente.

```python
while True:
    palabra_usuario = input("Ingresa una palabra en español (o escribe 'salir'): ")

    if palabra_usuario.lower() == "salir":
        print("Programa finalizado.")
        break

    print("Traducción:", traducir(palabra_usuario))
```
